```python
import numpy as np
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
```

```python
df = pq.read_table("part-00024-48f6ff92-4267-4a45-9460-515ecb3eb069.c000.snappy.parquet").to_pandas()
```

```python
df = df[['node_uuid', 'point_x', 'point_y', 'point_z', 'r', 'g', 'b']]
df = df[df['point_x'] != 0.0]
df
```

```python
pq.write_table(pa.Table.from_pandas(df), 'point_cloud.snappy.parquet', compression='snappy')
```
